```
import MeCab
m=MeCab.Tagger("mecabrc")
src='適当な日本語の文章を入れる（なんでもよい）'
res=m.parseToNode(src)

while res:
  features=res.feature.split(',')
  print(features[6],res.feature)
  res=res.next
```

```
import gensim
#ja.binは日本語で学習済みのモデル
model = gensim.models.Word2Vec.load('ja.bin')
print(len(model['機械']))
print(model['機械'])#←どの単語でもいい
```

```
model.most_similar('チョコレート')
```

```
#パリ-フランス+日本 を計算している
model.most_similar(positive=['パリ','日本'],negative=['フランス'])
```
